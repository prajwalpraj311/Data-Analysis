```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
day_inp = '20'
mon_inp = 'Apr'
```

```python
df = pd.read_csv('ODH Model pincode Tracker.csv')
dff1 = pd.read_excel("GTNL Report -"+day_inp+"th "+mon_inp+"'25.xlsx", sheet_name='Base Data for Work')
```

```python
df1 = df[['ODH/MDH Name','Type','Vendor','Zone','State']]
df1
```

```python
df2 = df1.drop_duplicates()
df2
```

```python
df2.rename(columns={'Vendor':'Vendor Name','ODH/MDH Name':'HubName'},inplace=True)
df2['HubName']= df2['HubName'].str.upper()
df2['Zone']=df2['Zone'].str.strip()
df2
```

```python
dff1.rename(columns={'CurrentHub':'HubName'},inplace=True)
dff1['HubName']= dff1['HubName'].str.upper()
dff1
```

```python
df3 = pd.merge(dff1, df2, on = "HubName", how = "left")
df3
```

```python
df4 = df3.dropna(subset = 'Vendor Name', how = 'all')
df4
```

```python
df5 = df4[df4["PRTO Reason"]!="MISSING_ITEM"]
df5
```

```python
def assign_bucket(return_amount):
    return ">=2k" if  return_amount >= 2000 else "<2k"
```

```python
df5["Bucket"] = df5["return_amount"].apply(assign_bucket)
df5
```

```python
df5.to_excel("gtnl_pendency"+day_inp+mon_inp+".xlsx")
```
