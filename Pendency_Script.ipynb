```python
def assign_shiptyp(row):
    shipment_id = str(row["ShipmentId"]) if pd.notna(row["ShipmentId"]) else ""
    rto_date = str(row["RTO DATE"]) if pd.notna(row["RTO DATE"]) else ""
    
    if ("R" in shipment_id) and ("-" in rto_date):
        return 'RVP'
    else:
        return 'RTO'

df1["Type Of Shipment"] = df1.apply(assign_shiptyp, axis=1)
```

```python
def latest_remarks(row):
    if row["New Latest Status"] in ["DELIVERED", "DELIVERY_UPDATE", "LOST", "RECEIVED_BY_EKL", "REJECTED_BY_SELLER","CLOSED","NOT_RECEIVED","DELIVERED_TO_VENDOR"]:
        return "Closed"
    elif ((row["New Latest Status"] == 'EXPECTED') or (row["New Latest Status"] =="UNTRACEABLE_BRSNR")) and (row["New BagStatus"] == "IN_TRANSIT"):
        return "LH bucket"
    elif (isinstance(row[" CurrentHub_y"], str) and ("MOTHERHUB" in row[" CurrentHub_y"].upper())):
        return "MH bucket"
    elif (isinstance(row[" CurrentHub_y"], str) and ((row[" CurrentHub_y"][-2:] == "01") or (row[" CurrentHub_y"][-3:] == "BTS"))):
        if row['New BagStatus']=="IN_TRANSIT":
            return "LH Pendency"
        elif row["New BagStatus"] == "REACHED":
            return "RC Pendency"
        else:
            return "BRSNR"
    elif (row["Type Of Shipment"] == 'RVP') and ((row["New Latest Status"] in ["EXPECTED","PICKUP_OUT_FOR_PICKUP"]) or ("NOTPICKED" in row["New Latest Status"])): 
        return "RVP"
    else:
        return "DH Pendency"

'''def latest_remarks(row):
    if row["New Latest Status"] in ["DELIVERED", "DELIVERY_UPDATE", "LOST", "RECEIVED_BY_EKL", "REJECTED_BY_SELLER","CLOSED","NOT_RECEIVED","DELIVERED_TO_VENDOR"]:
        return "Closed"
    elif ((row["New Latest Status"] == 'EXPECTED') or (row["New Latest Status"] =="UNTRACEABLE_BRSNR")) and (row["New BagStatus"] == "IN_TRANSIT"):
        return "LH bucket"
    elif (isinstance(row["Current Location"], str) and ("MOTHERHUB" in row["Current Location"].upper())):
        return "MH bucket"
    elif (isinstance(row["Current Location"], str) and ((row["Current Location"][-2:] == "01") or ("BTS" in row["Current Location"].upper()))):
        if row['New BagStatus']=="IN_TRANSIT":
            return "LH Pendency"
        elif row["New BagStatus"] == "REACHED":
            return "RC Pendency"
        else:
            return "BRSNR"
    else:
        return "DH Pendency"'''
```

```python
def assign_slab(days):
    if days > 7:
        return '>7'
    elif days > 5:
        return '5 to 7'
    elif days > 2:
        return '2 to 5'
    else:
        return '0 to 2'
```

```python
df2["Slab"] = df2["ageing"].apply(assign_slab)
df2.head()
```

```python
def assign_bucket( TotalPrice):
    return ">=2k" if  TotalPrice >= 2000 else "<2k"
```

```python
df2["Bucket"] = df2[" TotalPrice"].apply(assign_bucket)
df2
```
