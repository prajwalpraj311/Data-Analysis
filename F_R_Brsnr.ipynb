```python
import pandas as pd
import numpy as np
```

```python
Map = pd.read_csv("C:\\Users\\prajwal.d1\\Desktop\\Pendency\\ODH Model pincode Tracker.csv")
Map1 = Map[['ODH/MDH Name','Vendor']]
Map1.rename(columns={'Vendor':'Vendor Name','ODH/MDH Name':'HubName'},inplace=True)
Map1['HubName'] = Map1['HubName'].str.upper()
Map1 = Map1.drop_duplicates()
Map1
```

```python
df_brsnr = pd.read_excel("C:\\Users\\prajwal.d1\\Desktop\\Pendency\\BRSNR21A.xlsb",sheet_name="Raw Data")
```

```python
df_brsnr.columns
```

```python
df_brsnr = df_brsnr[df_brsnr["Aging Bucket"].isin([">30 days", "11-30 days", "3-5 days", "6-10 days"])]
df_brsnr
```

```python
df_brsnr = df_brsnr.drop(" Status", axis = "columns")
df_brsnr
```

```python
df_brsnr = df_brsnr[df_brsnr["Include or Exclude"] == "Include"]  # Keep only "Include" in Include/Exclude column
df_brsnr
```

```python
#Creating Fwd & Rvr BRSNR Databases
df_fbrsnr = df_brsnr
df_rbrsnr = df_brsnr


#Filtering Data in Fwd & Rvr BRSNR
df_fbrsnr = df_fbrsnr[df_fbrsnr["Review"].isin(["MH to ODH/MDH BRSNR"])]
df_rbrsnr = df_rbrsnr[df_rbrsnr["Review"].isin(["ODH/MDH to MH/RC BRSNR"])]
```

```python
df_fbrsnr
```

```python
df_rbrsnr
```

```python
# Fixing column name (removing leading space in "CurrentHub")
df_fbrsnr[" CurrentHub"] = df_fbrsnr[" CurrentHub"].str.upper()
df_rbrsnr["Source"] = df_rbrsnr["Source"].str.upper()
```

```python
# Merging with Map1
df_fbrsnr = pd.merge(df_fbrsnr, Map1, how="left", left_on=" CurrentHub", right_on="HubName")
df_fbrsnr["Type"] = "Forward"
df_rbrsnr = pd.merge(df_rbrsnr, Map1, how="left", left_on="Source", right_on="HubName")
df_rbrsnr["Type"] = "Reverse"
```

```python
df_fbrsnr = df_fbrsnr.dropna(subset = 'Vendor Name', how = 'all')
df_rbrsnr = df_rbrsnr.dropna(subset = 'Vendor Name', how = 'all')
```

```python
df_FinalBRSNR = pd.concat([df_fbrsnr, df_rbrsnr], ignore_index=True)
```

```python
df_FinalBRSNR
```

```python
df_FinalBRSNR.to_csv("BRSNR_Mulitracking.csv",index=False)
```
